```python
import copy
import numpy as np
def hueristic(source,target,n):
    x=[True for i in range(n) for j in range(n) if source[i][j]!=target[i][j] ]
    return len(x)
def swap(ti,tj,si,sj,temp):
    temp[ti][tj],temp[si][sj]=temp[si][sj],temp[ti][tj]
    return temp
def game():
    source=[[1,2,3],
      [5,7,8],
      [4,6,'_']]
    target=[[1,2,3],
       [4,5,6],
       [7,8,'_']]
    si,sj=2,2
    n=3
    val=n*n
    hueristic_list=[val,val,val,val]
    m=hueristic(source,target,n)
    counter=1
    tempt,tempb,templ,tempr=[],[],[],[]
    while m!=0 or m!=1:  # TOP
        old_source=copy.deepcopy(source)
        print("_____________________________________Iteration : ",counter,"__________________________________")
        print("Source : ",source)
        if si-1 >=0:
            ti,tj=si-1,sj
            tempt=swap(ti,tj,si,sj,copy.deepcopy(source))
            hueristic_list[0]=hueristic(tempt,target,3)
            print("TOP ",tempt, " hueristic : ",hueristic_list[0],"    swapped ",tempt[ti][tj],tempt[si][sj],"position : ",(ti,tj),(si,sj) )
        if sj-1>=0:    #LEFT
            li,lj=si,sj-1
            templ=source
            templ=swap(li,lj,si,sj,copy.deepcopy(source))
            hueristic_list[1]=hueristic(templ,target,3)
            print("LEFT ",templ, " hueristic : ",hueristic_list[1],"   **swapped ",templ[li][lj],templ[si][sj],"position : ",(li,lj),(si,sj))
        if si+1<n:    #BOTTOM
            bi,bj=si+1,sj
            tempb=source
            tempb=swap(bi,bj,si,sj,copy.deepcopy(source))
            hueristic_list[2]=hueristic(tempb,target,3)
            print("BOTTOM",tempb, " hueristic : ",hueristic_list[2],"    **swapped ",tempb[bi][bj],tempb[si][sj],"position : ",(bi,bj),(si,sj))
        if sj+1<n:    #RIGHT
            ri,rj=si,sj+1
            tempr=source
            tempb=swap(ri,rj,si,sj,copy.deepcopy(source))
            hueristic_list[3]=hueristic(tempr,target,3)
            print("RIGHT ",tempr, " hueristic : ",hueristic_list[3],"    **swapped ",tempr[ri][rj],tempr[si][sj],"position : ",(ri,rj),(si,sj))
        m=min(hueristic_list)
        hueristic_arr=np.array(hueristic_list)
        indexes=np.where(hueristic_arr==m)
        if old_source==source:
            min_=np.random.choice(indexes[0])
        else:
            min_=indexes[0][0]
        if min_==0:
            source=tempt
            si,sj=ti,tj
        elif min_==1:
            source=templ
            si,sj=li,lj
        elif min_==2:
            source=tempb
            si,sj=bi,bj
        elif min_==3:
            source=tempr
            si,sj=ri,rj
        else:
            pass
        counter+=1
        print("Selected source :",source)
        print("minimum hueristic ",m,"   Selected hueristic index ",min_,"\n\n")
```

```python
game()
```
